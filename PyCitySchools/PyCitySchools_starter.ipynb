```python
#***Bimi Bisht ***
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Load and read csv files
school_df = pd.read_csv("Resources/schools_complete.csv")
student_df = pd.read_csv("Resources/students_complete.csv")

```

```python
#Merging the two csv files and displaying data
Complete_data = pd.merge(student_df, school_df, how = "left", on = ["school_name", "school_name"])
Complete_data
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Create a dataframe to hold the above results

```python
#cCalculating the number of schools
school_count=Complete_data["school_name"].nunique()
school_count
```

```python
# Total student count
student_count=Complete_data["student_name"].count()
student_count
```

```python
# Calculating total budget
total_budget= Complete_data["budget"].unique().sum()
total_budget
```

```python
#  Calculating Average math score
avg_math_score = Complete_data["math_score"].mean()
avg_math_score
```

```python
# Calculating Average reading score
avg_reading_score = Complete_data["reading_score"].mean()
avg_reading_score
```

```python
# Calculating overall passing rate
overall_avg_score = (avg_reading_score + avg_math_score)/2
overall_avg_score
```

```python
# Calculating percent of students with 70 or higher  i.e with a passing math score
Complete_data["passing math"]= Complete_data["math_score"] >= 70
percent_passing_math = (Complete_data["passing math"]).mean()*100
percent_passing_math
```

```python
# Calculating percent of students with 70 or higher  i.e with a passing reading score
Complete_data["passing reading"]= student_df["reading_score"] >= 70
percent_passing_reading = (Complete_data["passing reading"]).mean()*100
percent_passing_reading
```

```python
#Calculating overall passing percent
overall_passing_rate = (avg_reading_score + avg_math_score)/2
overall_passing_rate
```

```python
# Creating dataframe
district_summary = pd.DataFrame({"Total Schools": [school_count],
                        "Total Students":[student_count],
                        "Total Budget": [total_budget],
                        "Average Reading Score": round(avg_reading_score,2),
                        "Average Math Score": round(avg_math_score,2),
                        "% Passing Math": round(percent_passing_math,2),
                         "% Passing Reading":round(percent_passing_reading,2),
# **Asked to calculate the overall passing rate by using instructions above ( It is overall passing rate not percent)                             
                         " Overall Passing Rate":round(overall_passing_rate,2)})

district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary[" Overall Passing Rate"] = district_summary[" Overall Passing Rate"].map("{:,.2f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Using merged data
Complete_data["passing math"] = Complete_data["math_score"] >= 70
Complete_data["passing reading"] = Complete_data["reading_score"] >= 70
Complete_data
```

```python
# groupby using school name
school_group = Complete_data.groupby(["school_name"]).mean()
school_group["Per Student Budget"] = school_group["budget"]/school_group["size"]
school_group["% Passing Math"] = round(school_group["passing math"]*100,2)
school_group["% Passing Reading"] = round(school_group["passing reading"]*100,2)
school_group["% Overall Passing Rate"] = round(((school_group["passing math"] + school_group["passing reading"])/2)*100,3)

# Merging with school_df to collect information about the type, size and budget
school_data = pd.merge(school_group, school_df, how="left", on=["school_name", "school_name"])
del school_data['size_y']
del school_data['budget_y']
del school_data['Student ID']
del school_data['School ID_x']

# Creating dataframe for the results
school_sum_df = pd.DataFrame({"School Name":  school_data["school_name"],
                                "School Type": school_data["type"],
                               "Total Students":school_data["size_x"],
                               "Total School Budget": school_data["budget_x"],
                               "Per Student Budget":school_data["Per Student Budget"], 
                               "Average Math Score":round(school_data["math_score"],2),
                               "Average Reading Score":round(school_data["reading_score"],2), 
                               "% Passing Math": school_data["% Passing Math"],
                               "% Passing Reading": school_data["% Passing Reading"],
                               "% Overall Passing Rate": school_data["% Overall Passing Rate"]}) 

#Formatting
school_sum_df["Total Students"] = school_sum_df["Total Students"].map("{:,.0f}".format)
school_sum_df["Total School Budget"] = school_sum_df["Total School Budget"].map("${:,.2f}".format)
school_sum_df["Per Student Budget"] = school_sum_df["Per Student Budget"].map("${:,.2f}".format)

#Display 
school_sum_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Finding top five performing schools
top_five_schools = school_sum_df.sort_values(["% Overall Passing Rate"],ascending= False)
top_five_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Creating a table to get math scores by grade level
ninthgrade = student_df.loc[student_df["grade"] == "9th"].groupby("school_name")["math_score"].mean()
tenthgrade = student_df.loc[student_df["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleventhgrade = student_df.loc[student_df["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelfthgrade = student_df.loc[student_df["grade"] == "12th"].groupby("school_name")["math_score"].mean()
math_score = pd.DataFrame({"9th": ninthgrade,
                        "10th": tenthgrade,
                        "11th": eleventhgrade,
                        "12th": twelfthgrade})
math_score = math_score[["9th","10th","11th","12th"]]
math_score.index.name =" "
math_score
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
#***Changed labels(group names) to creat NON overlapping ranges*** 
#---------------------------------------------------------------
# Creating bins for scores by school spending
spending_bins = [0.00, 584.99, 614.99, 644.99, 675]
group_names = ["<$585", "$585-614.99", "$615-644.99", "$645-675"]
school_data["Spending Ranges (Per Student)"] = pd.cut(school_data["Per Student Budget"], spending_bins, labels=group_names)
group_school_spending = school_data.groupby("Spending Ranges (Per Student)").mean() 
 

del group_school_spending['size_x']
del group_school_spending['budget_x']
del group_school_spending['Per Student Budget']
del group_school_spending['School ID_y']
del group_school_spending['passing math']
del group_school_spending['passing reading']



group_school_spending_renamed = group_school_spending.rename(columns={"reading_score": "Average Reading Score",
    "math_score": "Average Math Score"})
  
group_school_spending_renamed
```
